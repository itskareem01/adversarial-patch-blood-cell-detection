# 6. Evaluate And Optimise Thresholds for YOLOv10m

```python
import os
import numpy as np
from ultralytics import YOLO

# Load your trained YOLOv10m model
model = YOLO('/content/yolov10_bccd/baseline_m/weights/best.pt')

# ============================================
# 1. Set per‑class confidence thresholds (from Alam et al. 2021)
# ============================================
THRESHOLDS = {0: 0.55,   # RBC
              1: 0.35,   # WBC
              2: 0.25}   # Platelets

test_images_dir = '/content/BCCD_YOLO/images/test'
test_labels_dir = '/content/BCCD_YOLO/labels/test'

# Total counters
total_gt = {0:0, 1:0, 2:0}
total_pred = {0:0, 1:0, 2:0}

class_names = ['RBC', 'WBC', 'Platelets']

for img_file in os.listdir(test_images_dir):
    if not img_file.endswith('.jpg'):
        continue

    # ----- Ground truth -----
    label_path = os.path.join(test_labels_dir, img_file.replace('.jpg', '.txt'))
    gt_counts = {0:0, 1:0, 2:0}
    if os.path.exists(label_path):
        with open(label_path) as f:
            for line in f:
                cls = int(line.strip().split()[0])
                gt_counts[cls] += 1

    # ----- Prediction with manual confidence filter -----
    img_path = os.path.join(test_images_dir, img_file)
    preds = model(img_path, verbose=False)
    pred_counts = {0:0, 1:0, 2:0}

    if len(preds[0].boxes) > 0:
        boxes = preds[0].boxes
        cls = boxes.cls.cpu().numpy()
        conf = boxes.conf.cpu().numpy()

        # Apply per‑class thresholds
        for c in [0,1,2]:
            mask = (cls == c) & (conf > THRESHOLDS[c])
            pred_counts[c] = np.sum(mask)

    # ----- Update totals -----
    for c in [0,1,2]:
        total_gt[c] += gt_counts[c]
        total_pred[c] += pred_counts[c]

# ============================================
# 2. Print PAPER‑STYLE accuracy (total counts)
# ============================================
print("\n=== COUNTING ACCURACY (Total Counts, Paper Metric) ===")
for c, name in enumerate(class_names):
    gt = total_gt[c]
    pred = total_pred[c]
    if gt > 0:
        acc = (1 - abs(pred - gt) / gt) * 100
        print(f"{name:9s}: GT={gt:4d}, Pred={pred:4d}, Accuracy={acc:.2f}%")
    else:
        print(f"{name:9s}: No ground truth")
```

```python
import os
import numpy as np
from ultralytics import YOLO

model = YOLO('/content/yolov10_bccd/baseline_m/weights/best.pt')

# Fixed thresholds for WBC and PLT (from paper)
WBC_TH = 0.35
PLT_TH = 0.25

test_images_dir = '/content/BCCD_YOLO/images/test'
test_labels_dir = '/content/BCCD_YOLO/labels/test'

# Sweep RBC thresholds from 0.55 to 0.85
rbc_thresholds = [0.55, 0.60, 0.65, 0.70, 0.75, 0.80, 0.85]

print("RBC_Thresh\tGT_RBC\tPred_RBC\tAccuracy")
print("-" * 50)

for rbc_th in rbc_thresholds:
    total_gt = {0:0, 1:0, 2:0}
    total_pred = {0:0, 1:0, 2:0}

    for img_file in os.listdir(test_images_dir):
        if not img_file.endswith('.jpg'): continue

        # Ground truth
        label_path = os.path.join(test_labels_dir, img_file.replace('.jpg', '.txt'))
        gt_counts = {0:0,1:0,2:0}
        if os.path.exists(label_path):
            with open(label_path) as f:
                for line in f:
                    cls = int(line.strip().split()[0])
                    gt_counts[cls] += 1

        # Predict
        img_path = os.path.join(test_images_dir, img_file)
        preds = model(img_path, verbose=False)
        pred_counts = {0:0,1:0,2:0}
        if len(preds[0].boxes) > 0:
            boxes = preds[0].boxes
            cls = boxes.cls.cpu().numpy()
            conf = boxes.conf.cpu().numpy()

            # Apply per‑class thresholds with variable RBC threshold
            mask = (
                ((cls == 0) & (conf > rbc_th)) |
                ((cls == 1) & (conf > WBC_TH)) |
                ((cls == 2) & (conf > PLT_TH))
            )
            filtered_cls = cls[mask]
            unique, counts = np.unique(filtered_cls, return_counts=True)
            for u, c in zip(unique, counts):
                pred_counts[u] = c

        # Update totals
        for c in [0,1,2]:
            total_gt[c] += gt_counts[c]
            total_pred[c] += pred_counts[c]

    # Compute RBC accuracy
    gt_rbc = total_gt[0]
    pred_rbc = total_pred[0]
    if gt_rbc > 0:
        acc = (1 - abs(pred_rbc - gt_rbc) / gt_rbc) * 100
        print(f"{rbc_th:.2f}\t\t{gt_rbc}\t\t{pred_rbc}\t\t{acc:.2f}%")
```

```python
import os
import numpy as np
from ultralytics import YOLO
import json
from google.colab import drive

# ============================================
# 1. Load your trained YOLOv10m model
# ============================================
model = YOLO('/content/yolov10_bccd/baseline_m/weights/best.pt')  # adjust path if needed

# ============================================
# 2. Evaluation function (total‑count metric)
# ============================================
def evaluate_counting_accuracy(model, thresholds):
    """
    thresholds: dict {0: rbc_th, 1: wbc_th, 2: plt_th}
    Returns: dict with class names and accuracy, plus total_gt, total_pred
    """
    test_images_dir = '/content/BCCD_YOLO/images/test'
    test_labels_dir = '/content/BCCD_YOLO/labels/test'

    total_gt = {0:0, 1:0, 2:0}
    total_pred = {0:0, 1:0, 2:0}
    class_names = ['RBC', 'WBC', 'Platelets']

    for img_file in os.listdir(test_images_dir):
        if not img_file.endswith('.jpg'): continue
        label_path = os.path.join(test_labels_dir, img_file.replace('.jpg', '.txt'))
        gt_counts = {0:0,1:0,2:0}
        if os.path.exists(label_path):
            with open(label_path) as f:
                for line in f:
                    cls = int(line.strip().split()[0])
                    gt_counts[cls] += 1

        img_path = os.path.join(test_images_dir, img_file)
        preds = model(img_path, verbose=False)
        pred_counts = {0:0,1:0,2:0}
        if len(preds[0].boxes) > 0:
            boxes = preds[0].boxes
            cls = boxes.cls.cpu().numpy()
            conf = boxes.conf.cpu().numpy()
            for c in [0,1,2]:
                mask = (cls == c) & (conf > thresholds[c])
                pred_counts[c] = np.sum(mask)

        for c in [0,1,2]:
            total_gt[c] += gt_counts[c]
            total_pred[c] += pred_counts[c]

    accuracies = {}
    for c, name in enumerate(class_names):
        gt = total_gt[c]
        pred = total_pred[c]
        if gt > 0:
            acc = (1 - abs(pred - gt) / gt) * 100
            accuracies[name] = {'gt': int(gt), 'pred': int(pred), 'accuracy': round(acc, 2)}
        else:
            accuracies[name] = {'gt': int(gt), 'pred': int(pred), 'accuracy': None}
    return accuracies, total_gt, total_pred

# ============================================
# 3. Generic threshold sweep function
# ============================================
def sweep_threshold(model, class_to_sweep, threshold_range, fixed_thresholds):
    """
    class_to_sweep: 0=RBC, 1=WBC, 2=Platelets
    threshold_range: list of values to try
    fixed_thresholds: dict with thresholds for other classes
    """
    class_names = ['RBC', 'WBC', 'Platelets']
    sweep_name = class_names[class_to_sweep]

    print(f"\n=== SWEEPING {sweep_name} THRESHOLD ===")
    print(f"Fixed: RBC={fixed_thresholds.get(0, '?')}, WBC={fixed_thresholds.get(1, '?')}, PLT={fixed_thresholds.get(2, '?')}")
    print(f"{sweep_name}\tRBC_Acc\t\tWBC_Acc\t\tPLT_Acc\t\t{sweep_name}_GT/PRED")
    print("-" * 70)

    best_acc = 0
    best_th = None
    best_results = None

    for th in threshold_range:
        thresholds = fixed_thresholds.copy()
        thresholds[class_to_sweep] = th
        accuracies, _, _ = evaluate_counting_accuracy(model, thresholds)

        rbc_acc = accuracies['RBC']['accuracy']
        wbc_acc = accuracies['WBC']['accuracy']
        plt_acc = accuracies['Platelets']['accuracy']
        sweep_gt = accuracies[sweep_name]['gt']
        sweep_pred = accuracies[sweep_name]['pred']
        sweep_acc = accuracies[sweep_name]['accuracy']

        print(f"{th:.2f}\t{rbc_acc:.2f}%\t\t{wbc_acc:.2f}%\t\t{plt_acc:.2f}%\t\t{sweep_gt}/{sweep_pred} (acc={sweep_acc:.2f}%)")

        if sweep_acc > best_acc:
            best_acc = sweep_acc
            best_th = th
            best_results = (thresholds.copy(), accuracies)

    print("\n" + "="*50)
    print(f"✅ BEST {sweep_name} THRESHOLD: {best_th:.2f} (Accuracy: {best_acc:.2f}%)")
    print("="*50)
    return best_th, best_results
```

```python
# Fix WBC=0.35, PLT=0.25 (paper defaults), sweep RBC from 0.50 to 0.80
fixed = {1: 0.35, 2: 0.25}
rbc_range = [0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80]
best_rbc, _ = sweep_threshold(model, class_to_sweep=0, threshold_range=rbc_range, fixed_thresholds=fixed)
```

```python
import os
import numpy as np
from ultralytics import YOLO
import cv2
from PIL import Image

# Load your trained model
model = YOLO('/content/drive/MyDrive/blood_cell_project/best_yolov10m_bccd.pt')

# Paths for TXL-PBC
txl_test_images = '/content/TXL-PBC_Dataset/TXL-PBC/images/test'
txl_test_labels = '/content/TXL-PBC_Dataset/TXL-PBC/labels/test'

# Your model's class order: 0=RBC, 1=WBC, 2=Platelets
THRESHOLDS = {0: 0.65,   # RBC
              1: 0.50,   # WBC
              2: 0.25}   # Platelets

# Total counters
total_gt = {0:0, 1:0, 2:0}
total_pred = {0:0, 1:0, 2:0}
class_names = ['RBC', 'WBC', 'Platelets']

# Mapping from TXL class ID to your model class ID
txl_to_my = {
    0: 1,  # TXL WBC -> your WBC
    1: 0,  # TXL RBC -> your RBC
    2: 2   # TXL Platelets -> your Platelets
}

# Get all image files with common extensions
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp')
image_files = [f for f in os.listdir(txl_test_images) if f.lower().endswith(image_extensions)]

print(f"Found {len(image_files)} test images.")

for img_file in image_files:
    # Get base name without extension
    base_name = os.path.splitext(img_file)[0]
    label_path = os.path.join(txl_test_labels, base_name + '.txt')

    # Ground truth counts
    gt_counts = {0:0, 1:0, 2:0}
    if os.path.exists(label_path):
        with open(label_path) as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) < 5:
                    continue
                txl_cls = int(parts[0])
                if txl_cls in txl_to_my:
                    my_cls = txl_to_my[txl_cls]
                    gt_counts[my_cls] += 1

    # Predict with confidence filtering
    img_path = os.path.join(txl_test_images, img_file)
    preds = model(img_path, verbose=False)
    pred_counts = {0:0, 1:0, 2:0}

    if len(preds[0].boxes) > 0:
        boxes = preds[0].boxes
        cls = boxes.cls.cpu().numpy()
        conf = boxes.conf.cpu().numpy()

        for c in [0,1,2]:
            mask = (cls == c) & (conf > THRESHOLDS[c])
            pred_counts[c] = np.sum(mask)

    # Update totals
    for c in [0,1,2]:
        total_gt[c] += gt_counts[c]
        total_pred[c] += pred_counts[c]

# Calculate and display accuracies
print("\n=== TXL-PBC TEST SET RESULTS ===")
for c, name in enumerate(class_names):
    gt = total_gt[c]
    pred = total_pred[c]
    if gt > 0:
        acc = (1 - abs(pred - gt) / gt) * 100
        print(f"{name:9s}: GT={gt:4d}, Pred={pred:4d}, Accuracy={acc:.2f}%")
    else:
        print(f"{name:9s}: No ground truth (check data)")
```

```python
from ultralytics import YOLO
import os
import numpy as np

model = YOLO('/content/yolov10_combined/baseline_combined/weights/best.pt')
test_images = '/content/combined_dataset/images/test'
test_labels = '/content/combined_dataset/labels/test'

def evaluate(thresholds):
    total_gt = {0:0,1:0,2:0}
    total_pred = {0:0,1:0,2:0}
    for img_file in os.listdir(test_images):
        if not (img_file.endswith('.jpg') or img_file.endswith('.png')):
            continue
        # Ground truth
        label_file = os.path.splitext(img_file)[0] + '.txt'
        label_path = os.path.join(test_labels, label_file)
        gt_counts = {0:0,1:0,2:0}
        if os.path.exists(label_path):
            with open(label_path) as f:
                for line in f:
                    parts = line.strip().split()
                    if len(parts) >= 5:
                        cls = int(parts[0])
                        gt_counts[cls] += 1
        # Predict
        img_path = os.path.join(test_images, img_file)
        results = model(img_path, verbose=False)
        pred_counts = {0:0,1:0,2:0}
        if len(results[0].boxes) > 0:
            cls = results[0].boxes.cls.cpu().numpy()
            conf = results[0].boxes.conf.cpu().numpy()
            for c in [0,1,2]:
                mask = (cls == c) & (conf > thresholds[c])
                pred_counts[c] = np.sum(mask)
        for c in [0,1,2]:
            total_gt[c] += gt_counts[c]
            total_pred[c] += pred_counts[c]
    acc = {}
    for c in [0,1,2]:
        gt = total_gt[c]
        pred = total_pred[c]
        acc[c] = (1 - abs(pred - gt) / gt) * 100 if gt > 0 else 0
    return acc, total_gt, total_pred
```

```python
fixed = {1:0.50, 2:0.25}
best_rbc = 0.65  # starting point
best_acc = 0
print("Sweeping RBC threshold...")
for th in np.arange(0.60, 0.82, 0.02):
    thr = {0:th, **fixed}
    acc, _, _ = evaluate(thr)
    print(f"RBC={th:.2f} -> RBC Acc: {acc[0]:.2f}%, WBC Acc: {acc[1]:.2f}%, PLT Acc: {acc[2]:.2f}%")
    if acc[0] > best_acc:
        best_acc = acc[0]
        best_rbc = th
print(f"\nBest RBC threshold: {best_rbc:.2f} (RBC Acc: {best_acc:.2f}%)")
```

```python
fixed = {0:best_rbc, 2:0.25}
best_wbc = 0.50
best_acc = 0
print("\nSweeping WBC threshold...")
for th in np.arange(0.40, 0.62, 0.02):
    thr = {1:th, **fixed}
    acc, _, _ = evaluate(thr)
    print(f"WBC={th:.2f} -> RBC Acc: {acc[0]:.2f}%, WBC Acc: {acc[1]:.2f}%, PLT Acc: {acc[2]:.2f}%")
    if acc[1] > best_acc:
        best_acc = acc[1]
        best_wbc = th
print(f"\nBest WBC threshold: {best_wbc:.2f} (WBC Acc: {best_acc:.2f}%)")
```

```python
fixed = {0:best_rbc, 1:best_wbc}
best_plt = 0.25
best_acc = 0
print("\nSweeping Platelets threshold...")
for th in np.arange(0.40, 0.60, 0.02):
    thr = {2:th, **fixed}
    acc, _, _ = evaluate(thr)
    print(f"PLT={th:.2f} -> RBC Acc: {acc[0]:.2f}%, WBC Acc: {acc[1]:.2f}%, PLT Acc: {acc[2]:.2f}%")
    if acc[2] > best_acc:
        best_acc = acc[2]
        best_plt = th
print(f"\nBest Platelets threshold: {best_plt:.2f} (PLT Acc: {best_acc:.2f}%)")
```

```python
final_thr = {0:best_rbc, 1:best_wbc, 2:best_plt}
acc, total_gt, total_pred = evaluate(final_thr)
print("\n=== FINAL OPTIMISED RESULTS ===")
for c,name in enumerate(['RBC','WBC','Platelets']):
    print(f"{name:9s}: GT={total_gt[c]:4d}, Pred={total_pred[c]:4d}, Accuracy={acc[c]:.2f}%")
```
